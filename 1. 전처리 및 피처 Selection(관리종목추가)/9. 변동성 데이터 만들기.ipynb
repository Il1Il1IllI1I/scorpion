```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/5. label 지정 완료.csv",encoding='cp949')
df.columns
```

```python
fea = ['회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '소속코드', '산업명',
'유동자산증가율','총자본증가율',
  '자기자본순이익률', '총자본순이익률',
'자본분배율',  '부가가치율',
'순운전자본비율', '자기자본배율',
 "매입채무회전률",'매출채권회전기간']
```

```python
df = df[fea]
df
```

```python
for column in df.columns:
    if df[column].dtype == 'float64':  # 실수형 데이터만 처리
        # 비율 계산 (현재 값 / 이전 값)
        df[f'{column}_차이'] = df[column] - df[column].shift(1)

        # 변동률 계산 (현재 값 - 이전 값) / 이전 값
        df[f'{column}_변동률'] = (df[column] - df[column].shift(1)) / df[column].shift(1).abs()

# 결과 출력
df
```

```python
df.to_csv("./data/9. 변동률 데이터 추가.csv")
```
