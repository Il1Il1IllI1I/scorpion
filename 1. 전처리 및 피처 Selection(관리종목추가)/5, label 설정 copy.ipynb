# 관리 종목 지정 확인 하기

```python
df[df["회사명"].str.contains("폴리플러스")]
```

```python
df.reset_index(inplace=True)
df_label = df[['index','회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '소속코드']]
```

```python
df_label= df_label[df_label['소속코드']==4]
df_label= df_label.groupby('회사명')['index'].apply(list)
```

```python
df_label = pd.DataFrame(df_label)
df_label
```

```python
for j in df_label.index:
    for i in df.index:
        if df_label["index"][j][-1] == i:
            df.loc[i,'label'] = 1
df['label'] = df["label"].fillna(0)
```

```python
df_20121 = pd.read_csv("./data/2012년 관리종목 지정 종목 1.csv", encoding='cp949')
df_20122 = pd.read_csv("./data/2012년 관리종목 지정 종목 2.csv", encoding='cp949')
df_2012 = pd.concat([df_20121,df_20122])

df_all_2012 = df_2012["회사명"].unique().tolist()

len(df_all_2012)
```

```python
all_list = []
for i in df_all_2012:
    a = len(df[df["회사명"].str.contains(i)])
    if a >= 1 and  df[df["회사명"].str.contains(i)]["label"].sum()<1:
        b = df_2012[df_2012["회사명"]==i]
        all_list.append(b)

all_df = pd.concat(all_list)    
```

```python
all_df[all_df["공시제목"].str.contains("지정")]
```

```python
list_2012 = all_df[all_df["공시제목"].str.contains("지정")]["회사명"].unique().tolist()
list_2012
```

```python
index_list_2012 = []
for i in list_2012:
    num = df[df["회사명"].str.contains(i)].index[0]
    index_list_2012.append(num)
index_list_2012
```

```python
for i in index_list_2012:
    df.loc[i,"label"] = 1
```
