```python
import pandas as pd
import numpy as np

# 파일 로드
variables_df = pd.read_csv('변수12개.csv')

# 재무 비율 칼럼 선택 (label과 회계년도, 회사명, 거래소코드 제외)
financial_ratios = ['총자본순이익률', '매출액증가율', '자기자본순이익률', '매출액순이익률', '유동비율', 
                    '자기자본구성비율', '총자본회전률', '자기자본회전률', '총자본투자효율', '설비투자효율', '순이익증가율']

# 각 칼럼에 대한 파생변수 생성
for ratio in financial_ratios:
    # 이전 시점 대비 차이 및 비율
    variables_df[ratio + '_차이'] = variables_df.groupby('회사명')[ratio].diff()
    variables_df[ratio + '_비율'] = variables_df.groupby('회사명')[ratio].pct_change()

    # 전체 산업 대비 상위 10개 기업과의 차이 및 변동성
    top_10_mean = variables_df.groupby('회계년도')[ratio].nlargest(10).groupby('회계년도').mean()
    variables_df[ratio + '_상위10개_차이'] = variables_df[ratio] - variables_df['회계년도'].map(top_10_mean)
    variables_df[ratio + '_상위10개_변동성'] = variables_df[ratio + '_상위10개_차이'].abs()

    # 전체 산업 대비 평균과의 차이 및 변동성
    industry_mean = variables_df.groupby('회계년도')[ratio].mean()
    variables_df[ratio + '_산업평균_차이'] = variables_df[ratio] - variables_df['회계년도'].map(industry_mean)
    variables_df[ratio + '_산업평균_변동성'] = variables_df[ratio + '_산업평균_차이'].abs()

# 파생변수가 포함된 데이터프레임을 CSV 파일로 저장
save_path = '파생변수.csv'
variables_df.to_csv(save_path, index=False)
```
