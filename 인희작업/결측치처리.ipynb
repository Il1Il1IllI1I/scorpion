```python
import pandas as pd
import numpy as np
```

```python
# 초기 재무비율로만 구성된 데이터 불러오기 (재무비율 10개 + 총자본이익률 + 라벨)
raw = pd.read_csv("변수12개.csv")
# 변동성지표 만들 때 0의 값은 문제가 될 수 있으므로 미리 개수 파악.
zero_counts = raw.eq(0).sum()
print(zero_counts)
# 순이익증가율에서 0의 값이 많이 보임. 파악할 필요 있음.
```

```python
#결측치 확인
raw.isnull().sum()
```

```python
# 파생변수 생성 후 완성 데이터 불러오기 (파생변수 만드는 코드는 따로 정리할것.)
df = pd.read_csv("파생변수real.csv",index_col=0)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
# 2010년 데이터를 구할 수 없어서, 2011년에 결측치 발생. 수축기를 2012년부터 보기로함.
df['회계년도'] = pd.to_datetime(df['회계년도'])
df = df[df['회계년도'].dt.year != 2011]
```

```python
# 순이익증가율_비율 결측치를 0으로 대체
df['순이익증가율_비율'] = df['순이익증가율_비율'].fillna('0')
```

```python
#결측치 확인
missing_values = df.isnull().sum()
# 결측치가 있는 컬럼만 선택
missing_values[missing_values > 0]
# 역시나 순이익증가율_비율 값에서 결측치가 많이 나옴 -> 원데이터에 0의 값이 많아서 비율로 계산할 때 결측치로 뱉음.
```
