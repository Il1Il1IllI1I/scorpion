```python
import pandas as pd
import numpy as np
```

```python
# 데이터셋 분할
# Re-loading the provided dataset due to the previous exception
file_path = '데이터전처리완료.csv'
data = pd.read_csv(file_path)

# 1. Sorting the data by '회계년도' in descending order
data_sorted = data.sort_values(by='회계년도', ascending=False)

# 2. Extracting data for 2012-2013 for train_data
train_data = data_sorted[data_sorted['회계년도'].str.contains('2012|2013')]
train_data_file = 'train_data.csv'
train_data.to_csv(train_data_file, index=False)

# 3. Extracting data for 2014-2017 for 확장기_test
expansion_test_data = data_sorted[data_sorted['회계년도'].str.contains('2014|2015|2016|2017')]
expansion_test_file = 'test_확장기.csv'
expansion_test_data.to_csv(expansion_test_file, index=False)

# 4. Extracting data for 2018-2019 for 수축기_test
contraction_test_data = data_sorted[data_sorted['회계년도'].str.contains('2018|2019')]
contraction_test_file = 'test_수축기.csv'
contraction_test_data.to_csv(contraction_test_file, index=False)

train_data_file, expansion_test_file, contraction_test_file
```
